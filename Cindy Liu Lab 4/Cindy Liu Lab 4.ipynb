### Scrape the IMDB website of the top 100 movies. Isolate the title, the date of release, and one other element of your choosing. Put the data into a csv. 

```python
from bs4 import BeautifulSoup
import requests
import html5lib
import csv
import pandas as pd
```

```python
source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(source, "lxml")

csv_file = open("imdb_scrape.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Release Year", "Runtime"])

for element in soup.find_all("div", class_="lister-item-content"):
    
    title = element.h3.a.text.strip()
    release = element.find("span", class_="lister-item-year text-muted unbold").text.strip()
    runtime = element.find("span", class_="runtime").text.strip()

    csv_writer.writerow([title, release, runtime])
    
csv_file.close()
```
