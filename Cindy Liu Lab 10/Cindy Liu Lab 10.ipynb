GUI 1: 
Using the quotes API, build a GUI that provides a drop-down menu of people randomly chosen from the API.
I want a user to be able to pick from that list and receive a quote that corresponds to that person.
Ideally, I want to be able to reset that menu & allow people to choose from a new list.

```python
from tkinter import *
import requests
import json

root = Tk()
root.title("Quotes")
root.geometry("650x600")

text_box = Text(root, height=20, width=50)
text_box.pack()
people = []

for i in range(4):
    response = requests.get("https://api.quotable.io/random")
    data = response.json()
    people.append(data["author"])

item_list = people

def choice():
    text_box.delete(0.0, END)
    response = requests.get("https://api.quotable.io/random")
    data = response.json()
    
    quote = data["content"]
    person = data["author"]
    person = value.get()
    
    label = Label(root, text=f"The author {person} has a quote, {quote}.")
    text_box.insert(END, quote)
    
value = StringVar()
value.set("Choose a person")

dropdown = OptionMenu(root, value, *item_list)
choose = Button(root, text="Click to choose", command=choice)

dropdown.pack()
choose.pack()

root.mainloop()
    
```
