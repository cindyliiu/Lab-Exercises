Use our weather data we covered in Monday’s class and convert those three elements (day, temp, description) into a CSV. Using the techniques covered in this slide deck, convert that CSV into a bar graph – y axis should be the temperature, x axis should be the day.

```python
import requests
import json
import csv
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

lat = "42.0894"
lon = "-75.9695"
genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")

json_file = genius.json()

forecast = json_file["properties"]["forecast"]
new_request = requests.get(forecast)

json = new_request.json()

dict = json["properties"]["periods"]

data_file = open("weather_data.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(data_file)
csv_writer.writerow(["Name", "Temp", "Detail"])

for i in dict:
    name = i["name"]
    temp = i["temperature"]
    detail = i["detailedForecast"]
    print(name)
    print(temp)
    print(detail)
    
    csv_writer.writerow([name, temp, detail])
    
data_file.close()
```

```python
sheet = pd.read_csv("weather_data.csv")
sheet
```

```python
#source: https://www.tutorialspoint.com/increasing-the-space-for-x-axis-labels-in-matplotlib

from matplotlib import pyplot as plt

df = pd.DataFrame(sheet)

day = df["Name"]
temp = df["Temp"]

fig = plt.figure(figsize = (18,7))
plt.bar(day[0:10], temp[0:10])

plt.xlabel("Day")
plt.ylabel("Temp")
plt.title("Temperature of Each Day")
spacing = 0.002
fig.subplots_adjust(bottom=spacing)

plt.show()
```

Then I want you to build a program that gets public holiday info from 10 countries of your choosing in an efficient way.  Count the total number of holidays and store that info. Be careful, some countries might not have holidays listed in the API!

```python
import requests
import json
import csv
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
total = 0
countries = ["US", "CH", "HT", "MT", "VA", "VE", "CO", "ES", "MX", "NZ"]

for country in countries:
    response = requests.get(f'https://date.nager.at/api/v3/PublicHolidays/2023/{country}')
    
    if response.status_code == 200:
        holiday_count = len(response.json())
        total += holiday_count
        print(f"{country}: {holiday_count}")
    else:
        print("None")
        
print(f"\nTotal Number of Holidays: {total}")
```

Bonus: Create a bar graph to visualize the number of public holidays in the 10 countries you counted. 

```python
x_axis = countries
y_axis = [13, 19, 20, 14, 15, 38, 20, 34, 7, 23]

plt.bar(x_axis, y_axis)
plt.title("Public Holidays")
plt.xlabel("Countries")
plt.ylabel("Holiday Count")
plt.show()
```
